# Task 4 - Grouping

## a. gender

```python
df_gender = df[['customerID','gender']].groupby(by='gender').count()
df_gender
```

```python
## sort results from the largest number
df_gender.sort_values('customerID', ascending = False)
```

## b. InternetService

```python
df_InternetService = df[['customerID','InternetService']].groupby(by='InternetService').count()
df_InternetService
```

```python
## sort results from the largest number
df_InternetService.sort_values('customerID', ascending = False)
```

## c. PaymentMethod

```python
df_PaymentMethod = df[['customerID','PaymentMethod']].groupby(by='PaymentMethod').count()
df_PaymentMethod
```

```python
## sort results from the largest number
df_PaymentMethod.sort_values('customerID', ascending = False)
```

```python
df['ChargesType'] = pd.cut(df['MonthlyCharges'],
                            bins=[0,35.5,70.5,89.9,99999],
                            labels=['Low','Medium','High','Very High'])
```
